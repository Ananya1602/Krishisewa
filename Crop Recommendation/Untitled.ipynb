change upper case to lower case and remove spaces to get same names

```python
def change_name(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
crop['label']=crop['label'].apply(change_name)
fert['Crop']=fert['Crop'].apply(change_name)
```

Check if changes made succesfully

```python
crop['label'].unique()
```

```python
fert['Crop'].unique()
```

```python
# chickpeas,mungbeans, lentils, pigeonpeas,mothbeans names to be changed in fert dataset
```

```python
fert['Crop'] = fert['Crop'].replace('mungbeans','mungbean')
fert['Crop'] = fert['Crop'].replace('lentils(masoordal)','lentil')
fert['Crop'] = fert['Crop'].replace('pigeonpeas(toordal)','pigeonpeas')
fert['Crop'] = fert['Crop'].replace('mothbean(matki)','mothbeans')
fert['Crop'] = fert['Crop'].replace('chickpeas(channa)','chickpea')
```

```python
crop_name = crop['label'].unique()
crop_name_fert =fert['Crop'].unique()
crop_name_fert
```

```python
common_name = []
for i in crop_name_fert:
    if i in crop_name:
        common_name.append(i)
```
